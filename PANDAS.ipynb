```python
from sklearn.linear_model import LinearRegression # LinearRegression IS A FUNCTION THAT IS USED TO CREATE A MODEL WITH ONLY GIVEN NUMERIC VALUES THAT CALCULATES THE GIVEN VALUES AND TO GIVE THE OUTPUT
```

```python
dfsport=fiatdf.loc[fiatdf['model']=='sport']
print(dfsport.to_string())
```

```python
X1 = fiatdf.drop(columns=['ID','engine_power','lat','lon','price','model'])
Y1 = fiatdf['price']
print(X1.to_string())
print(Y1.to_string())
```

```python
n=int(input())
for i in range(n):
    a=input("MODEL :")
    b=fiatdf.loc[fiatdf['model']==a]
    print(b.to_string())
```

```python
mod=LinearRegression()  # USING THE LINEARREGRESSION A MODEL TO GIVE THE OUTPUT IS CREATED.                
mod.fit(X1,Y1)          # FITTING THE X, Y DATAFRAMES IN THE MODEL OF LINEARREGRESSION.
```

```python
ans = mod.predict([[2190,70000,1]])   # PREDICTING OUTPUT BY GIVING THE VALUES TO THE PREDICTION FROM THE MODEL. 
print(ans)
```

```python
ans1 = mod.predict([[2190,70000,1]])
print(ans1)
```

```python
n=int(input())
for i in range(1,n+1):
    print("DETAILS OF CAR,",i,":")
    b,c,d=map(int,input().split())
    ans2 = mod.predict([[b,c,d]])
    print(ans2)
```
